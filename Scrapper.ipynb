# Amazon Scrapper


### Requirements

```python
import csv
from bs4 import BeautifulSoup
import pandas as pd
```

```python
from selenium import webdriver
```

## Starting Web Driver

```python
driver = webdriver.Chrome()
```

## Collection

```python
pages = range(1, 168)
df = pd.DataFrame()
failed_list = []
page_cnt = 1
for page in pages:
    link = "https://www.amazon.com/s?i=computers-intl-ship&bbn=16225007011&rh=n%3A16225007011%2Cn%3A13896617011%2Cn%3A565108&dc&page={}".format(page)
    driver.get(link)
    soup = BeautifulSoup(driver.page_source, 'html.parser')
    results = soup.find_all('a', {"class": "a-link-normal a-text-normal"}, href=True)
    urls = []

    for html in results:
        href = html.get('href')
        href = 'https://www.amazon.com/' + href
        urls.append(href)
    cnt = 1
    mx = len(urls)
    for url in urls:
        attrs = {}
        attrs_failed = {}
        failed_str = ''
        driver.get(url)
        soup = BeautifulSoup(driver.page_source, 'html.parser')
        try:
            results = soup.find('span', {"id": "priceblock_ourprice"})
            attrs['price'] = results.text
        except:
            failed_str += 'price, '
        try:
            results = soup.find('span', {"class": "a-icon-alt"})
            attrs['rating'] = results.text[:3]
        except:
            failed_str += 'rating, '
        data = []
        data2 = []
        try:
            table = soup.find('table', attrs={'id':'productDetails_techSpec_section_1'})
            table_body = table.find('tbody')

            rows = table_body.find_all('tr')
            for row in rows:
                key = row.find_all('th')
                value = row.find_all('td')
                keys = [ele.text.strip() for ele in key]
                values = [ele.text.strip() for ele in value]
                data.append(keys[0])
                data2.append(values[0][1:])


            for i in range(len(data)):
                attrs[data[i]] = data2[i]
        except:
            failed_str += 'table_1, '
        data = []
        data2 = []
        try:
            table = soup.find('table', attrs={'id':'productDetails_techSpec_section_2'})
            table_body = table.find('tbody')

            rows = table_body.find_all('tr')
            for row in rows:
                key = row.find_all('th')
                value = row.find_all('td')
                keys = [ele.text.strip() for ele in key]
                values = [ele.text.strip() for ele in value]
                data.append(keys[0])
                data2.append(values[0][1:])

            for i in range(len(data)):
                attrs[data[i]] = data2[i]
        except:
            failed_str += 'table_2,'
        print('added item', cnt, ' out of ', mx, '(page {} out of 168)'.format(page_cnt))

        cnt += 1
        df = df.append(attrs, ignore_index=True)
        if failed_str != '':
             failed_list.append([failed_str, url])
    print(len(failed_list))
    page_cnt += 1
```
